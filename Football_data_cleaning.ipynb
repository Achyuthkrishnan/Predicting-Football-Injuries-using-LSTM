# Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("player_injuries.csv")
```

```python
df.head(10)
```

```python
df.shape
```

##### Removing entires having unknown injury.

```python
df = df[df['injury_reason'].str.lower() != 'unknown injury']
```

```python
df.shape
```

##### Removing entries not having end date

```python
df = df.dropna(subset=['end_date'])
```

```python
df.shape
```

##### Set days_missed as y

```python
y = df['days_missed']
```

```python
y.shape
```

##### Drop target from features

```python
X = df.drop(columns=['days_missed'])
```

```python
X.shape
```

```python
df = X.copy()
```

```python
df['y'] = y
```
