# Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/achyuth/DL_Project/Predicting-Football-Injuries-using-LSTM/player_injuries.csv")
```

```python
df.head(10)
```

```python
df.shape
```

##### Removing entires having unknown injury.

```python
unwanted_injuries = ['unknown injury', 'ill', 'cyst on jaw', 'influenza', 'fitness', 'corona virus', 'cold', 'appendectomy', 'intestinal virus', 'stomach flu', 'flu', 'rest']   
```

```python
for injury in unwanted_injuries:
    df = df[df['injury_reason'].str.lower() != injury]
```

```python
df.shape
```

##### Removing entries not having end date

```python
df = df.dropna(subset=['end_date'])
```

```python
df.to_csv("player_injuries_clean_v2.csv", index=False)
```
