# Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
injuries = pd.read_csv("/home/achyuth/DL_Project/Predicting-Football-Injuries-using-LSTM/player_injuries.csv")
profiles=pd.read_csv("/home/achyuth/DL_Project/Predicting-Football-Injuries-using-LSTM/player_profiles.csv")
```

```python
profiles_small = profiles[["player_id", "date_of_birth"]]

df = injuries.merge(profiles_small, on="player_id", how="left")
```

```python
df.head(10)
```

```python
df.shape
```

##### Removing entires having unknown injury.

```python
unwanted_injuries = ['unknown injury', 'ill', 'cyst on jaw', 'influenza', 'fitness', 'corona virus', 'cold', 'appendectomy', 'intestinal virus', 'stomach flu', 'flu', 'rest']   
```

```python
for injury in unwanted_injuries:
    df = df[df['injury_reason'].str.lower() != injury]
```

```python
df.shape
```

##### Removing entries not having end date

```python
df = df.dropna(subset=['end_date'])
```

```python
df.shape
```

```python
#converting to date-time
df["date_of_birth"] = pd.to_datetime(df["date_of_birth"])
df["from_date"] = pd.to_datetime(df["from_date"])

#computing age of player at time of injury
df["age_at_injury"] = (
    (df["from_date"] - df["date_of_birth"]).dt.days // 365.25
)
```
